```python
import pandas as pd
brazil_weather = pd.DataFrame({
    "city": ["Natal", "Recife", "João Pessoa"],
    "temperature": [32,45,30],
    "humidity": [80,60,78]
})
brazil_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["New York", "Chicago", "Orlando"],
    "temperature": [21,14,35],
    "humidity": [68,65,75]
})
us_weather
```

```python
df = pd.concat([brazil_weather, us_weather])
df
```

```python
df = pd.concat([brazil_weather, us_weather], ignore_index=True)
df
```

```python
df = pd.concat([brazil_weather, us_weather], keys=["Brazil", "US"])
df
```

```python
df.loc["US"]
```

```python
s = pd.Series([7,12,15], name="windspeed")
s
df = pd.concat([brazil_weather, s], axis=1)
df
```

```python
df1 = pd.DataFrame({
    "city": ["New York", "Chicago", "Orlando"],
    "temperature": [21,14,35],
})
df2 = pd.DataFrame({
    "city": ["Chicago", "Orlando", "New York"],
    "humidity": [65,68,75],
})
df = pd.merge(df1, df2,on="city")
df
```

```python
df1 = pd.DataFrame({
    "city": ["New York", "Chicago", "Orlando", "Baltimore"],
    "temperature": [21,14,35, 25],
})
df2 = pd.DataFrame({
    "city": ["Chicago", "New York", "San Francisco"],
    "humidity": [65,68,75],
})
df = pd.merge(df1, df2,on="city", how="outer")
df
```

```python
df1 = pd.DataFrame({
    "city": ["New York", "Chicago", "Orlando", "Baltimore"],
    "temperature": [21,14,35, 25],
})
df2 = pd.DataFrame({
    "city": ["Chicago", "New York", "San Francisco"],
    "humidity": [65,68,75],
})
df = pd.merge(df1, df2,on="city", how="inner")
df
```

```python
df1 = pd.DataFrame({
    "city": ["New York", "Chicago", "Orlando", "Baltimore"],
    "temperature": [21,14,35, 25],
})
df2 = pd.DataFrame({
    "city": ["Chicago", "New York", "San Francisco"],
    "humidity": [65,68,75],
})
df = pd.merge(df1, df2,on="city", how="left")
df
```

```python
df1 = pd.DataFrame({
    "city": ["New York", "Chicago", "Orlando", "Baltimore"],
    "temperature": [21,14,35, 25],
})
df2 = pd.DataFrame({
    "city": ["Chicago", "New York", "San Francisco"],
    "humidity": [65,68,75],
})
df = pd.merge(df1, df2,on="city", how="right")
df
```

```python
df1 = pd.DataFrame({
    "city": ["New York", "Chicago", "Orlando", "Baltimore"],
    "temperature": [21,14,35, 25],
})
df2 = pd.DataFrame({
    "city": ["Chicago", "New York", "San Francisco"],
    "humidity": [65,68,75],
})
df = pd.merge(df1, df2,on="city", how="outer", indicator=True)
df
```
